```python
import csv
import json
import pandas as pd
from nltk.probability import FreqDist
from pprint import pprint
import nltk
from nltk.tokenize import word_tokenize
from textblob import TextBlob
from textblob.classifiers import NaiveBayesClassifier
```

```python
data=[]
with open("Tweets_EN_sentiment_train.csv", encoding="utf-8") as csvfile:
    reader = csv.DictReader(csvfile )
    for d in reader:
        data.append((d["text"],d["class"])) 
data
```

```python
text = data[1]
text
```

```python
cl = NaiveBayesClassifier(data)
```

```python
cl.classify(text)
```

## Tokinization 

```python
#token = nltk.word_tokenize(data)
tokenized_sents = [word_tokenize(i) for i in data]
for i in tokenized_sents:
    print(i) 
```

### analisar as frequencias das Palavras

```python
fdist = FreqDist()
```

```python
for word in tokenized_sents:
    fdist[str(word)]+=1
fdist
    
```

```python
top_20= fdist.most_common(20)
top_20
```

```python
from nltk.tokenize import blankline_tokenize

blank = blankline_tokenize(fdist)

len(blank)
```
