```python
##LIBRARIES

import pandas as pd
import csv
#create a graphics 
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
import csv
from textblob import TextBlob
#!pip install vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

## Upload Tweet File

```python
## UPLOAD TRAIN DATA SET 

tweeter_train = []
with open("Tweets_EN_sentiment_train.csv", encoding="utf-8") as csvfile:
    reader = csv.DictReader(csvfile )
    for d in reader:
        tweeter_train.append((d["text"])) 


## UPLOAD TESTE DATA SET 

tweeter_test = []
with open("Tweets_EN_sentiment_test.csv", encoding="utf-8") as csvfile:
    reader = csv.DictReader(csvfile )
    for d in reader:
        tweeter_test.append((d["text"])) 
        
print('Example of a sentence in train data set: ', tweeter_train[1])        
print('Example of a sentence in test data set: ', tweeter_test[1])
```

## Preparation of the data - transform a sentiment string into a numeric value

```python
#CONVERT STRING INTO A NUMBER IN TRAIN DATA SET 

#transform the twetters classification in number in tweeter train date set 
text = []
for i in tweeter_train:
    if i == 'pos':
        text.append(1)
    else:
        text.append(-1)
        
#CONVERT STRING INTO A NUMBER IN TRAIN DATA SET 
       
#transform the twetters classification in number in tweeter train date set 
text_test = []
for i in tweeter_test:
    if i == 'pos':
        text_test.append(1)
    else:
        text_test.append(-1)
```

## VADER sentiment Clasifier - Train set

```python
analyzer_sentiment = []
analyzer = SentimentIntensityAnalyzer()

def Sentiment_Intensity_Analyzer(tweeter_train):
    for t in tweeter_train:
        analyze = analyzer.polarity_scores(t)
        analyzer_sentiment.append(analyze)
    #print(analyzer_sentiment)

Sentiment_Intensity_Analyzer(tweeter_train)
analyzer_sentiment
```

```python
analyzer_p = []
resul_s = []

for p in range(len(analyzer_sentiment)):
    if analyzer_sentiment[p]['pos'] > 0:
        analyzer_p.append(1)
    else: 
        analyzer_p.append(-1)
   
```

```python
#accuracy 
counter = 0
for t in range(len(text)):
    if text[t] == analyzer_p[t]:
        counter = counter + 1
        
print('Numbers of right classifications: ', counter )   
print('Accuracy: ', counter/len(tweeter_train)*100, '%' )  
```
